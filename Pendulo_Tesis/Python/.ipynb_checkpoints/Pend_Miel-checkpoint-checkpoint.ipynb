```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

# function that returns dy/dt
def pend_miel(x,t,gamma):
    dxdt = gamma - np.sin(x)
    return dxdt

# initial condition
x0 = 100

# time points
t = np.linspace(0,50,999)

# solve ODEs
gamma = 1.1
x1 = odeint(pend_miel,x0,t,args=(gamma,))
gamma = 100
x2 = odeint(pend_miel,x0,t,args=(gamma,))
gamma = 2
x3 = odeint(pend_miel,x0,t,args=(gamma,))


# plot results
plt.plot(t,x1,'r-',linewidth=2,label='$\gamma=1.1$')
plt.plot(t,x2,'b--',linewidth=2,label='$\gamma=100$')
plt.plot(t,x3,'g:',linewidth=2,label='$\gamma=2$')
plt.xlabel('time')
plt.ylabel('$x(t)$')
plt.grid(True)
plt.legend()
plt.show()
```

```python
plt.plot(t,np.sin(x1),'r',linewidth=2,label='$\gamma=1.1$')
plt.xlabel('time')
plt.ylabel('$sin(x)$')
plt.grid(True)
plt.legend()
plt.show()
```

```python
plt.plot(t,np.sin(x2),'b',linewidth=2,label='$\gamma=100$')
plt.xlabel('time')
plt.ylabel('$sin(x)$')
plt.grid(True)
plt.legend()
plt.show()
```

```python
plt.plot(t,np.sin(x3),'g',linewidth=2,label='$\gamma=2$')
plt.xlabel('time')
plt.ylabel('$sin(x)$')
plt.grid(True)
plt.legend()
plt.show()
```

```python
plt.plot(t,pend_miel(x2,t,100),'b',linewidth=2,label='$\gamma=100$')
plt.xlabel('time')
plt.ylabel('$\dot{\Theta}$')
plt.ylim((0,200))
plt.grid(True)
plt.legend()
plt.show()
```

```python
plt.plot(t,pend_miel(x3,t,100),'g',linewidth=2,label='$\gamma=2$')
plt.xlabel('time')
plt.ylabel('$\dot{\Theta}$')
plt.ylim((0,200))
plt.grid(True)
plt.legend()
plt.show()
```
