```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

# function that returns dy/dt
def pend_miel(x,t,gamma):
    dxdt = gamma - np.sin(x)
    return dxdt

# initial condition
#x0 = 5

# time points
#t = np.linspace(0,50,999)

# solve ODEs
#gamma = 1.1
#x1 = odeint(pend_miel,x0,t,args=(gamma,))
#gamma = 100
#x2 = odeint(pend_miel,100,t,args=(gamma,))
#gamma = 2
#x3 = odeint(pend_miel,x0,t,args=(gamma,))

# plot results
#plt.plot(t,x1,'r-',linewidth=2,label='$\gamma=1.1$')
#plt.plot(t,x2,'b--',linewidth=2,label='$\gamma=100$')
#plt.plot(t,x3,'g:',linewidth=2,label='$\gamma=2$')
#plt.xlabel('time')
#plt.ylabel('$x(t)$')
#plt.grid(True)
#plt.legend()
#plt.show()
```

```python
#Parametro
gamma = 100
#C.I.
x0 = 0
#Intervalo de tiempo
t1 = np.linspace(0,50,500)
#Solucion de EDO
x1 = odeint(pend_miel,x0,t1,args=(gamma,))


plt.plot(t1,pend_miel(x1,t1,gamma),'b',linewidth=2,label='$\gamma=100$')
plt.xlabel('time')
plt.ylabel('$\dot{\Theta}$')
plt.ylim((0,150))
plt.grid(True)
plt.legend()
plt.show()

plt.plot(t1,np.sin(x1),'b',linewidth=2,label='$\gamma=100$')
plt.xlabel('time')
plt.ylabel('$\sin(\Theta(t))$')
plt.grid(True)
plt.legend()
plt.show()

plt.plot(t1,x1,'b',linewidth=2,label='$\gamma=100$')
plt.xlabel('time')
plt.ylabel('$\Theta(t)$')
plt.grid(True)
plt.legend()
plt.show()
```

Análisis del péndulo $\dot{\theta}=\gamma + sin(\theta)$

```python
def cerocl(gamma,x):
    return gamma - np.sin(x)
gamma = 0.5
x = np.linspace(-np.pi,np.pi)

# setting the axes at the centre
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

#Ticks
plt.xticks([-np.pi, -np.pi/2, np.pi/2, np.pi],
       [r'$-\pi$', r'$-\pi/2$',  r'$+\pi/2$', r'$+\pi$'])

plt.yticks([gamma],
       [r'$\gamma$'])
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')



# plot the function
plt.plot(x,cerocl(gamma,x))
plt.xlabel('$\Theta(t)$',horizontalalignment='right', x=1.0)
plt.ylabel('$\dot{\Theta}(t)$',horizontalalignment='right', y=1.0)

plt.grid(0)

for label in ax.get_xticklabels() + ax.get_yticklabels():
    label.set_fontsize(13)
    label.set_bbox(dict(facecolor='white', edgecolor='None', alpha=0.65 ))



# show the plot
fig.savefig("gamma1_Pend_Miel.png",dpi=72)
plt.show()

```
