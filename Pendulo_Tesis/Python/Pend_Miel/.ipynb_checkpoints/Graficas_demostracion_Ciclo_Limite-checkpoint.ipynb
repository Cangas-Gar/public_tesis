```python
import numpy as np
import matplotlib.pyplot as plt

# Create a new figure of size 8x5 points, using 80 dots per inch
fig, ax = plt.subplots(figsize=(8,5), dpi=80)

# setting the axes at the centre:
# making the top and right spine invisible:
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

ax.spines['left'].set_position(('axes',0.))
ax.spines['bottom'].set_position(('axes',0.))


#endpoint includes the last interval point, pi in this case
x = np.linspace(2, 4.2, 256,endpoint=True)

def P1(X):
    return (X - 3) - (X - 3)**3/6 + 3

plt.xticks([2, 2.676227,3.4215, 4.2],
       [r'$v_1$',r'$\widetilde{v}_1$', r'$\widetilde{v}_2$',r'$v_2$'])
ax.xaxis.set_ticks_position('bottom')

plt.yticks([], [])



ax.plot(x, P1(x) , color="black", linewidth=2.5, linestyle="-",  label=r'$P(v)$')
ax.plot(x, x , color="black", linewidth=1, linestyle="--",  label=r'$v$')

plt.xlim((1.8,4.5))

plt.xlabel('$v(0)$',fontsize=12, horizontalalignment='right', x=1.0)
plt.ylabel('$v(2\pi)$',fontsize=12, horizontalalignment='right', y=1.0)
ax.legend(loc='upper left', frameon=False)

for label in ax.get_xticklabels() + ax.get_yticklabels():
    label.set_fontsize(13)
    label.set_bbox(dict(facecolor='white', edgecolor='None', alpha=0.65 ))

fig.savefig("Pinc_MapVARIOS.png",dpi=72)
plt.show()

#############################################
#############################################

# Create a new figure of size 8x5 points, using 80 dots per inch
fig, ax = plt.subplots(figsize=(8,5), dpi=80)

# setting the axes at the centre:
# making the top and right spine invisible:
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

ax.spines['left'].set_position(('axes',0.))
ax.spines['bottom'].set_position(('axes',0.))


#endpoint includes the last interval point, pi in this case
x = np.linspace(2, 4.2, 256,endpoint=True)

def P2(X):
    return np.sqrt(X)+1.5

plt.xticks([2, 3.32288, 4.2],
       [r'$v_1$',r'$\widetilde{v}$', r'$v_2$'])

plt.yticks([],
       [])
ax.xaxis.set_ticks_position('bottom')


ax.plot(x, P2(x) , color="black", linewidth=2.5, linestyle="-",  label=r'$P(v)$')
ax.plot(x, x , color="black", linewidth=1, linestyle="--",  label=r'$v$')

plt.xlim((1.8,4.5))

plt.xlabel('$v(0)$',fontsize=12, horizontalalignment='right', x=1.0)
plt.ylabel('$v(2\pi)$',fontsize=12, horizontalalignment='right', y=1.0)
ax.legend(loc='upper left', frameon=False)

for label in ax.get_xticklabels() + ax.get_yticklabels():
    label.set_fontsize(13)
    label.set_bbox(dict(facecolor='white', edgecolor='None', alpha=0.65 ))

fig.savefig("Pinc_MapUNICO.png",dpi=72)




plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8,5), dpi=80)


# setting the axes at the centre:
# making the top and right spine invisible:
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

ax.spines['left'].set_position(('axes',0.))
ax.spines['bottom'].set_position(('axes',0.))




def P1(x):
    return 0.5 - 0.2*np.cos(x)

def P2(x):
    return 0.5 + 0.2*np.cos(x)
    
z = np.pi/2    #punto de interseccion de las funciones que defin

x = np.linspace(0, 2.5, 100,endpoint=True)


plt.xticks([0, z, 2.5],
       [r'$0$', r'$z$', r'$2\pi$'])
ax.xaxis.set_ticks_position('bottom')
plt.yticks([P1(0), P2(0)], [r'$x$', r'$y$'])






x0 = 2.5
y1 = P2(2.5)
plt.annotate(r"$P(y)$",xy=(x0,y1),xycoords='data',xytext=(11,-5),
                  textcoords='offset points', fontsize=14)


y0 = P1(2.5)
plt.annotate(r"$P(x)$",xy=(x0,y0),xycoords='data',xytext=(11,-5),
                  textcoords='offset points', fontsize=14)

#Graficar la lnea vertical punteada
plt.plot([x0,x0],[1,0],'k--',linewidth=1.0)
plt.plot([z,z],[P2(z),0],'k:',linewidth=1.0)


ax.plot(x, P1(x) , color="black", linewidth=2.5, linestyle="-")
ax.plot(x, P2(x) , color="black", linewidth=2.5, linestyle="-")
plt.xlim((0,3))
plt.ylim((0,1))
plt.xlabel('$\Theta$',fontsize=12, horizontalalignment='right', x=1.0)
plt.ylabel('$v$',fontsize=12, horizontalalignment='right', y=1.0)


for label in ax.get_xticklabels() + ax.get_yticklabels():
    label.set_fontsize(14)
    label.set_bbox(dict(facecolor='white', edgecolor='None', alpha=0.65 ))
    
plt.savefig("Monotonia_Poinc.png",dpi=73)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8,5), dpi=80)


# setting the axes at the centre:
# making the top and right spine invisible:
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.spines['left'].set_position(('data',0.))
ax.spines['bottom'].set_position(('axes',0.))

def cerocl(x, beta, b):
    return (beta - np.sin(x))/2*b

def v1(x,beta, b):
    return ((beta - 1)/2*b) - 0.05

def v2(x,beta, b):
    return ((beta + 1)/2*b) + 0.05

x = np.linspace(0, 2*np.pi, 100,endpoint=True)

ax.plot(x, cerocl(x, 1.5,0.2) , color="black", linewidth=1.5, linestyle="-")
plt.xlabel('$\Theta$',fontsize=12, horizontalalignment='right', x=1.0)
plt.ylabel('$v$',fontsize=12, horizontalalignment='right', y=1.0)
ax.hlines(y=v1(x, 1.5,0.2), xmin=0, xmax=2*np.pi + 0.4, linewidth=1.5, color = "black", linestyle="--")
ax.hlines(y=v2(x, 1.5,0.2), xmin=0, xmax=2*np.pi +0.4,  linewidth=1.5, color = "black", linestyle="--")


plt.xticks([0, 2*np.pi],
       [r'$0$',  r'$2\pi$'])
ax.xaxis.set_ticks_position('bottom')
plt.yticks([],[])

xc = 2*np.pi
yc = cerocl(2*np.pi, 1.5,0.2)
plt.annotate(r'$v_c$',xy=(xc,yc),xycoords='data',xytext=(7,-5),
                  textcoords='offset points', fontsize=14)


x1 = 2*np.pi + 0.4
y1 = v1(x, 1.5,0.2)
plt.annotate(r'$v=v_1$',xy=(x1,y1),xycoords='data',xytext=(11,-5),
                  textcoords='offset points', fontsize=14)

x2 = 2*np.pi +0.4
y2 = v2(x, 1.5,0.2)
plt.annotate(r'$v=v_2$',xy=(x2,y2),xycoords='data',xytext=(11,-5),
                  textcoords='offset points', fontsize=14)

x3 = 2*np.pi
y3 = v2(x, 1.5,0.2)
ax.plot([x3],[v2(x, 1.5,0.2)], 'ko')
plt.annotate(r'$A^\prime$',xy=(x3,y3),xycoords='data',xytext=(7,5),
                  textcoords='offset points', fontsize=14)

x4 = 0
y4 = v2(x, 1.5,0.2)
ax.plot([x4],[v2(x, 1.5,0.2)], 'ko')
plt.annotate(r'$A$',xy=(x4,y4),xycoords='data',xytext=(7,5),
                  textcoords='offset points', fontsize=14)

x5 = 2*np.pi
y5 = v1(x, 1.5,0.2)
ax.plot([x5],[v1(x, 1.5,0.2)], 'ko')
plt.annotate(r'$B^\prime$',xy=(x5,y5),xycoords='data',xytext=(7,5),
                  textcoords='offset points', fontsize=14)

x6 = 0
y6 = v1(x, 1.5,0.2)
ax.plot([x6],[v1(x, 1.5,0.2)], 'ko')
plt.annotate(r'$B$',xy=(x6,y6),xycoords='data',xytext=(7,5),
                  textcoords='offset points', fontsize=14)

plt.plot([x3,x3],[.33,-0.03],'k--',linewidth=1.5)

#Flechas para arriba
x_direct = 1
y_direct = 1
y_pos = y1-0.019
x_pos = np.arange(1,6,1.5)
for i in x_pos:
    ax.quiver(i,y_pos,x_direct,y_direct, scale_units='xy', scale=15)
#Flechas para abajo   
x_direct = 1
y_direct = -1
y_pos = y2+0.0199
x_pos = np.arange(1,6,1.5)
for i in x_pos:
    ax.quiver(i,y_pos,x_direct,y_direct, scale = 15)
#Flechas horizontales
x_direct = 1
y_direct = 0
x_pos = np.arange(1,6,1.5)
for i in x_pos:
    ax.quiver(i,cerocl(i, 1.5,0.2),x_direct,y_direct, scale = 10)

for label in ax.get_xticklabels() + ax.get_yticklabels():
    label.set_fontsize(13)

plt.savefig("RectanguloABA'B'.png",dpi=73)
plt.show()
```
