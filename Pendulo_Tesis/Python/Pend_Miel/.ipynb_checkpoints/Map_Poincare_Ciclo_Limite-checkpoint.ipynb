```python
import numpy as np
import matplotlib.pyplot as plt

# Create a new figure of size 8x5 points, using 80 dots per inch
fig, ax = plt.subplots(figsize=(8,5), dpi=80)

# setting the axes at the centre:
# making the top and right spine invisible:
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

ax.spines['left'].set_position(('axes',0.))
ax.spines['bottom'].set_position(('axes',0.))


#endpoint includes the last interval point, pi in this case
x = np.linspace(2, 4.2, 256,endpoint=True)

def P1(X):
    return (X - 3) - (X - 3)**3/6 + 3

plt.xticks([2, 2.676227,3.4215, 4.2],
       [r'$v_1$',r'$\widetilde{v}_1$', r'$\widetilde{v}_2$',r'$v_2$'])
ax.xaxis.set_ticks_position('bottom')

plt.yticks([], [])



ax.plot(x, P1(x) , color="black", linewidth=2.5, linestyle="-",  label=r'$P(v)$')
ax.plot(x, x , color="black", linewidth=1, linestyle="--",  label=r'$v$')

plt.xlim((1.8,4.5))

plt.xlabel('$v(0)$',fontsize=12, horizontalalignment='right', x=1.0)
plt.ylabel('$v(2\pi)$',fontsize=12, horizontalalignment='right', y=1.0)
ax.legend(loc='upper left', frameon=False)

for label in ax.get_xticklabels() + ax.get_yticklabels():
    label.set_fontsize(13)
    label.set_bbox(dict(facecolor='white', edgecolor='None', alpha=0.65 ))

fig.savefig("Pinc_MapVARIOS.png",dpi=72)
plt.show()

#############################################
#############################################

# Create a new figure of size 8x5 points, using 80 dots per inch
fig, ax = plt.subplots(figsize=(8,5), dpi=80)

# setting the axes at the centre:
# making the top and right spine invisible:
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

ax.spines['left'].set_position(('axes',0.))
ax.spines['bottom'].set_position(('axes',0.))


#endpoint includes the last interval point, pi in this case
x = np.linspace(2, 4.2, 256,endpoint=True)

def P2(X):
    return np.sqrt(X)+1.5

plt.xticks([2, 3.32288, 4.2],
       [r'$v_1$',r'$\widetilde{v}$', r'$v_2$'])

plt.yticks([],
       [])
ax.xaxis.set_ticks_position('bottom')


ax.plot(x, P2(x) , color="black", linewidth=2.5, linestyle="-",  label=r'$P(v)$')
ax.plot(x, x , color="black", linewidth=1, linestyle="--",  label=r'$v$')

plt.xlim((1.8,4.5))

plt.xlabel('$v(0)$',fontsize=12, horizontalalignment='right', x=1.0)
plt.ylabel('$v(2\pi)$',fontsize=12, horizontalalignment='right', y=1.0)
ax.legend(loc='upper left', frameon=False)

for label in ax.get_xticklabels() + ax.get_yticklabels():
    label.set_fontsize(13)
    label.set_bbox(dict(facecolor='white', edgecolor='None', alpha=0.65 ))

fig.savefig("Pinc_MapUNICO.png",dpi=72)




plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8,5), dpi=80)

# setting the axes at the centre:
# making the top and right spine invisible:
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

ax.spines['left'].set_position(('axes',0.))
ax.spines['bottom'].set_position(('axes',0.))


x = np.linspace(0, 2.5, 100,endpoint=True)


plt.xticks([0, 1.5708, 2.5],
       [r'$0$', r'$z$', r'$2\pi$'])
ax.xaxis.set_ticks_position('bottom')
plt.yticks([0.5 - 0.36*np.cos(0), 0.5 + 0.36*np.cos(0)], [r'$x$', r'$y$'])






x0 = 2.5
y1 = 0.5 + 0.36*np.cos(2.5)
plt.annotate(r"$P(y)$",xy=(x0,y1),xycoords='data',xytext=(11,-5),
                  textcoords='offset points', fontsize=14)


y0 = 0.5 - 0.36*np.cos(2.5)
plt.annotate(r"$P(x)$",xy=(x0,y0),xycoords='data',xytext=(11,-5),
                  textcoords='offset points', fontsize=14)

#Graficar la línea vertical punteada
plt.plot([x0,x0],[1,0],'k--',linewidth=1.0)
plt.plot([1.57,1.570],[0.5 + 0.36*np.cos(1.5708),0],'k:',linewidth=1.0)


ax.plot(x, 0.5 - 0.36*np.cos(x) , color="black", linewidth=2.5, linestyle="-")
ax.plot(x, 0.5 + 0.36*np.cos(x) , color="black", linewidth=2.5, linestyle="-")
plt.xlim((0,3))
plt.ylim((0,1))
plt.xlabel('$\Theta$',fontsize=12, horizontalalignment='right', x=1.0)
plt.ylabel('$v$',fontsize=12, horizontalalignment='right', y=1.0)


for label in ax.get_xticklabels() + ax.get_yticklabels():
    label.set_fontsize(14)
    label.set_bbox(dict(facecolor='white', edgecolor='None', alpha=0.65 ))
    
plt.savefig("Monotonia_Poinc.png",dpi=73)
plt.show()
```
